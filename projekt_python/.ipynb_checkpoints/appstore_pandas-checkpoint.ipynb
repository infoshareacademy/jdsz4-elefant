```python
def rank_price (cena):
    if cena >= 3:
        return "Above 3$"
    else:
        return "Below 3$"
    
```

```python
%matplotlib notebook
plt.style.use("ggplot")
cm = sns.light_palette("green", as_cmap=True)
wykres_cena = df["Price"].apply(rank_price).value_counts().plot(kind='pie',autopct='%.0f',cmap="viridis", label = None,fontsize=15)  # ile ponad śrenia ile poniżej
```

```python
cena = wykres_cena.get_figure()
print(cena)
cena.savefig(os.path.join(os.path.abspath(""), "cena.png"))
# nazwa2.savefig(os.path.join(os.path.abspath(""), "nazwa2.png"))
```

```python
def rank_ocena (ocena):
    if ocena >= 4:
        return "Above 4"
    else:
        return "Below 4"
```

```python
%matplotlib notebook
plt.style.use("ggplot")
cm = sns.light_palette("green", as_cmap=True)
wykres_ocena = df["Average User Rating"].apply(rank_ocena).value_counts().plot(kind='pie',autopct='%.0f', cmap="viridis", label = None,fontsize=15)
```

```python
ocena = wykres_ocena.get_figure()
print(ocena)
ocena.savefig(os.path.join(os.path.abspath(""), "ocena.png"))
# nazwa2.savefig(os.path.join(os.path.abspath(""), "nazwa2.png"))
```

```python
def rank_wiek (wiek):
    if wiek >= 46:
        return "Above 9"
    else:
        return "Below 9"
```

```python
def trans_age(x):
    if x == '4+':
        return 4
    if x == '9+':
        return 9
    if x == '12+':
        return 12
    if x == '17+':
        return 17
df['Trans_Age'] = df['Age Rating'].apply(trans_age)
```

```python
%matplotlib notebook
plt.style.use("ggplot")
cm = sns.light_palette("green", as_cmap=True)
wykres_wiek = df["Age Rating"].apply(rank_wiek).value_counts().plot(kind='pie',autopct='%.0f',cmap="viridis", label = None,fontsize=15)
```
