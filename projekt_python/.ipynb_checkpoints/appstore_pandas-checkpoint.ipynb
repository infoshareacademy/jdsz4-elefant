```python
def rank_price (cena):
    if cena >= 0.8134187128304518:
        return "Above Average"
    else:
        return "Below Average"
    
```

```python
plt.style.use("ggplot")
df["Price"].apply(rank_price).value_counts().plot(kind='pie', legend = True)  # ile ponad śrenia ile poniżej
```
