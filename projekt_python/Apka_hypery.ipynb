```python
import tkinter as tk
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats as st
import pandas as pd
import seaborn as sns
from PIL import Image, ImageTk
import random
from tkinter import font  as tkfont
import requests
from io import BytesIO
import webbrowser
from tkHyperlinkManager import HyperlinkManager
```

```python
df= pd.read_csv('appstore_games.csv')
size_frame = df.sort_values(by=['Average User Rating','User Rating Count'],ascending=False)
```

```python
class SampleApp(tk.Tk):

    def __init__(self, *args, **kwargs):
        tk.Tk.__init__(self, *args, **kwargs)

        self.title_font = tkfont.Font(family='Helvetica', size=18, weight="bold", slant="italic")
        self.geometry('2000x2000')
        
        
        container = tk.Frame(self)
        container.pack(side="top", fill="both", expand=True)
        container.grid_rowconfigure(0, weight=1)
        container.grid_columnconfigure(0, weight=1)
        
        self.frames = {}
        for F in (StartPage, PageOne, PageTwo):
            page_name = F.__name__
            frame = F(parent=container, controller=self)
            self.frames[page_name] = frame

            frame.grid(row=0, column=0, sticky="nsew")

        self.show_frame("StartPage")

    def show_frame(self, page_name):
        '''Show a frame for the given page name'''
        frame = self.frames[page_name]
        frame.tkraise()
    

class StartPage(tk.Frame):

    def __init__(self, parent, controller):
        tk.Frame.__init__(self, parent)
        self.controller = controller
        self.configure(background='white')
        label1 = tk.Label(self, text='Welcome to App Store Games Recommender by Elefant',bg='white', font=('Times New Roman',40))
        label1.grid(column=2,row=1)
        
        
        button1 = tk.Button(self, text="Games recommender by parameters",bg='gold',font=('Times New Roman',16,'bold'),height=5,width=35,
                            command=lambda: controller.show_frame("PageOne"))
        button2 = tk.Button(self, text="Games recommender by title ",bg='gold',font=('Times New Roman',16,'bold'),height=5,width=35,
                            command=lambda: controller.show_frame("PageTwo"))
        button1.grid(column=2,row=5)
        button2.grid(column=2,row=6)
        
        image = Image.open('apple_im1.jpg')
        image = image.resize((600, 500), Image.ANTIALIAS)
        photo = ImageTk.PhotoImage(image)

        labeli = tk.Label(self,image=photo)
        labeli.image = photo
        labeli.grid(column=2,row=30,rowspan=12)
        
        

class PageOne(tk.Frame):

    def __init__(self, parent, controller):
        tk.Frame.__init__(self, parent)
        self.controller = controller
        self.configure(background='white')
        label1 = tk.Label(self, text="Recommend by parameters",bg='white', font=('Times New Roman',30))
        label1.grid(column=2,row=1)
        lable2=tk.Label(self,text='Enter your free space (in Mb)',font=('Times New Roman',20),bg='white')
        lable2.grid(column=2,row=2)
        lable3=tk.Label(self,text='Enter your age',font=('Times New Roman',20),bg='white')
        lable3.grid(column=2,row=4)
        lable4=tk.Label(self,text='Enter price of the game',font=('Times New Roman',20),bg='white')
        lable4.grid(column=2,row=6)
        lable5=tk.Label(self,text='Enter language of the game',font=('Times New Roman',20),bg='white')
        lable5.grid(column=2,row=8)
        lable5=tk.Label(self,text='Enter genre of the game',font=('Times New Roman',20),bg='white')
        lable5.grid(column=2,row=10)
        self.entry_field1 = tk.Entry(self,font=('Ubuntu', 24))
        self.entry_field1.grid(column=2,row=3)
        self.entry_field2 = tk.Entry(self,font=('Ubuntu', 24))
        self.entry_field2.grid(column=2,row=5)
        self.entry_field3 = tk.Entry(self,font=('Ubuntu', 24))
        self.entry_field3.grid(column=2,row=7)
        self.entry_field4 = tk.Entry(self,font=('Ubuntu', 24))
        self.entry_field4.grid(column=2,row=9)
        self.entry_field5 = tk.Entry(self,font=('Ubuntu', 24))
        self.entry_field5.grid(column=2,row=11)
        
        button1 = tk.Button(self, text = "Go to the start page",bg = 'gold',font=('Times New Roman',16,'bold'),height=5,width=30,command = lambda: controller.show_frame("StartPage"))
        button1.grid(column=2,row=30)
        button2 = tk.Button(self, text = "Recommend",bg = 'gold',font=('Times New Roman',16,'bold'),height=5,width=30,command =self.display)
        button2.grid(column=2,row=20)
        
        image = Image.open('apple_im1.jpg')
        image = image.resize((600, 500), Image.ANTIALIAS)
        photo = ImageTk.PhotoImage(image)

        labeli = tk.Label(self,image=photo)
        labeli.image = photo
        labeli.grid(column=1,row=1,rowspan=12)
        
        labeli2 = tk.Label(self,image=photo)
        labeli2.image = photo
        labeli2.grid(column=3,row=1,rowspan=12)


    def display(self):
        apple=self.Best_game()

        games_display = tk.Text(master=self,height=10,width=50,font=('Times New Roman',16))
        hyperlink = HyperlinkManager(games_display)
        games_display.grid(column=2,row=25,)

        games_display.insert(tk.END,apple)
    
    def Best_game(self):
        try:
            size=float(self.entry_field1.get())
        except:
            size=10000
        try:
            age=float(self.entry_field2.get())
        except:
            age=4
        try:
            price=float(self.entry_field3.get())
        except:
            price=0
        try:
            lang=str(self.entry_field4.get()).upper()
        except:
            lang='EN'
        try:
            genre=str(self.entry_field5.get()).capitalize()
        except:
            genre='Strategy'
        size_frame = df.sort_values(by=['Average User Rating','User Rating Count'],ascending=False)
        size*=1000000
        sf=size_frame.loc[size_frame['Size']<size]
        lf=sf[sf['Languages'].str.contains(lang,na=False)]
        if age <4:
            return 'To young to play'
        elif age<9:
            agerat=['4+']
        elif age<12:
            agerat=['4+','9+']
        elif age<17:
            agerat=['4+','9+','12+']
        else:
            agerat=['4+','9+','12+','17+']
        af=lf[lf['Age Rating'].str.contains('|'.join(agerat))]
        pf=af.loc[af['Price']<=price]
        gf=pf[pf['Genres'].str.contains(genre,na=False)]
        xf=gf[['ID']]
        xf=xf.head(30)
        xf_list=xf['ID'].tolist()
        if xf_list==[]:
            return 'No game available with this set of parameters'
        rf1=random.choice(xf_list)
        rf2=random.choice(xf_list)
        while  rf2==rf1:
            rf2=random.choice(xf_list)
        rf3=random.choice(xf_list)
        while   rf3==rf1 or rf3==rf2:
            rf3=random.choice(xf_list)
        ost_name1=df['Name'].where(df['ID']==rf1).dropna().item()
        ost_name2=df['Name'].where(df['ID']==rf2).dropna().item()
        ost_name3=df['Name'].where(df['ID']==rf3).dropna().item()
        ost_url1=df['URL'].where(df['ID']==rf1).dropna().item()
        ost_url2=df['URL'].where(df['ID']==rf2).dropna().item()
        ost_url3=df['URL'].where(df['ID']==rf3).dropna().item()
        hiper1=webbrowser.open(ost_url1)
        hiper2=webbrowser.open(ost_url2)
        hiper3=webbrowser.open(ost_url3)
#         ost_url2=df['URL'].where(df['ID']==rf2).dropna().item()
#         ost_url3=df['URL'].where(df['ID']==rf3).dropna().item()
        icon_url1=df['Icon URL'].where(df['ID']==rf3).dropna().item()
        response1 = requests.get(icon_url1)
        img1 = Image.open(BytesIO(response1.content))
        img1.save("out.jpg", "JPEG", quality=80, optimize=True, progressive=True)
        image = Image.open('out.jpg')
        image = image.resize((10, 10), Image.ANTIALIAS)
        photo = ImageTk.PhotoImage(image)
        return f'*** {ost_name1}\n*** {ost_name2}\n*** {ost_name3}                                                                                                                                                                                                                                                                                                                                                                                                                                                                                     {hiper1}{hiper2}{hiper3}'
    


class PageTwo(tk.Frame):

    def __init__(self, parent, controller):
        tk.Frame.__init__(self, parent)
        self.controller = controller
        
        label1 = tk.Label(self, text="Recommend by title",bg='white', font=('Times New Roman',30))
        label1.grid(column=2,row=1)
        lable2=tk.Label(self,text='Enter title of a game',font=('Times New Roman',20),bg='white')
        lable2.grid(column=2,row=2)
        self.configure(background='white')
        button1 = tk.Button(self, text="Go to the start page",bg='gold',font=('Times New Roman',16,'bold'),height=5,width=30,
                           command=lambda: controller.show_frame("StartPage"))
        button1.grid(column=2,row=30)
        button2 = tk.Button(self, text = "Recommend",bg = 'gold',font=('Times New Roman',16,'bold'),height=5,width=30,command =self.display)
        button2.grid(column=2,row=10)
        
        self.entry_field1 = tk.Entry(self,font=('Ubuntu', 24))
        self.entry_field1.grid(column=2,row=3)
        
        image = Image.open('apple_im1.jpg')
        image = image.resize((600, 500), Image.ANTIALIAS)
        photo = ImageTk.PhotoImage(image)

        labeli = tk.Label(self,image=photo)
        labeli.image = photo
        labeli.grid(column=1,row=1,rowspan=12)
        
        labeli2 = tk.Label(self,image=photo)
        labeli2.image = photo
        labeli2.grid(column=3,row=1,rowspan=12)
        
    def display(self):
        apple=self.Other_game()

        games_display = tk.Text(master=self,height=10,width=50,font=('Times New Roman',16))
        games_display.grid(column=2,row=25,)

        games_display.insert(tk.END,apple)   
        
        
    def Other_game(self):
        title=str(self.entry_field1.get()).lower()
        size_frame = df.sort_values(by=['Average User Rating','User Rating Count'],ascending=False)
            #title=str(title)
        size_frame['Namel']=size_frame['Name'].str.lower()
        tit=size_frame.loc[size_frame['Namel']==(title)]
        tit_price=float(tit['Price'])
        tit_size=float(tit['Size'])
            #tit_genre=tit['Genres']
        tit_avrrat=float(tit['Average User Rating'])
        tit_ratcount=float(tit['User Rating Count'])
            #tit_inpurch=tit['In-app Purchases']
        tit_age=str(tit['Age Rating'].item())
        sf=size_frame.loc[size_frame['Size']<(1.5*tit_size)]
        if tit_age=='4+':
            agerat=['4+']
        elif tit_age=='9+':
            agerat=['4+','9+']
        elif tit_age=='12+':
            agerat=['4+','9+','12+']
        elif tit_age=='17+':
            agerat=['4+','9+','12+','17+']
        af=sf[sf['Age Rating'].str.contains('|'.join(agerat))]
        pf=af.loc[af['Price']<=tit_price]
        avf=pf.loc[pf['Average User Rating']>=(tit_avrrat)]
        rcf=avf.loc[pf['User Rating Count']>=(tit_ratcount)]
        xf=rcf.head(30)
        xf_list=xf['Name'].tolist()
        rf1=random.choice(xf_list)
        while rf1==title:
            rf1=random.choice(xf_list)
        rf2=random.choice(xf_list)
        while rf2==title or rf2==rf1:
            rf2=random.choice(xf_list)
        rf3=random.choice(xf_list)
        while rf3==title or rf3==rf1 or rf3==rf2:
            rf3=random.choice(xf_list)
        
        return f'The first recomended game for you is {rf1}\nThe second recomended game for you is {rf2}\nThe third recomended game for you is {rf3}'

        
if __name__ == "__main__":
    app = SampleApp()
    app.mainloop()
```
